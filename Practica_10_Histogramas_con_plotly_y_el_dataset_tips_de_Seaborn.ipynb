```python
import seaborn as sns

tips_df = sns.load_dataset("tips")
display(tips_df.head())
```

```python
import plotly.express as px

fig = px.histogram(tips_df, x="tip")
fig.show()
```

```python
import plotly.express as px

fig = px.histogram(tips_df, x="tip", nbins=20, color="day")
fig.show()
```
