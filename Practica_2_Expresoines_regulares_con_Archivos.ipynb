## Associate values with companies

### Subtask:
Create a data structure (e.g., a dictionary or list of tuples) to store the "Valor actual:" values associated with each company name. Since the data appears sequentially in the file (company name followed by its value), we can likely pair them up based on their order of appearance.


**Reasoning**:
The task requires extracting pairs of company names and their corresponding "Valor actual" values from the file. This involves iterating through the file, identifying lines with company names and value, and creating a structured list of these pairs.



```python
import re

# Re-open the file to ensure reading from the beginning
main = open('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/informe_acciones.txt', encoding='latin-1')

# Initialize an empty list to store (company name, valor actual) pairs
company_data = []

# Initialize variables to store the last found company name and to indicate if we are looking for a value
last_company_name = None

# Regex patterns
company_pattern = re.compile(r'^Asunto:\s*(.+)')
valor_pattern = re.compile(r'Valor actual:\s*\$([\d.]+)')

for line in main:
    line = line.rstrip()

    # Check for company name
    company_match = company_pattern.search(line)
    if company_match:
        last_company_name = company_match.group(1).strip()
        continue # Move to the next line after finding a company name

    # Check for "Valor actual:"
    valor_match = valor_pattern.search(line)
    if valor_match and last_company_name:
        valor_actual = valor_match.group(1)
        # Pair the last found company name with the current valor actual
        company_data.append((last_company_name, valor_actual))
        last_company_name = None # Reset after pairing

# Close the file
main.close()

# Print the extracted pairs
print("Extracted Company Data (Name, Valor Actual):")
for item in company_data:
    print(item)
```

## Convert values to numbers

### Subtask:
Convert the extracted "Valor actual:" values from strings to a numerical type (like float) so that calculations can be performed.


**Reasoning**:
Iterate through the `company_data` list and convert the second element of each tuple (the "Valor actual:" value) from a string to a float.



```python
for i in range(len(company_data)):
    company_data[i] = (company_data[i][0], float(company_data[i][1]))

print("Converted Company Data (Name, Valor Actual as Float):")
for item in company_data:
    print(item)
```

## Calculate statistics per company

### Subtask:
Iterate through the data structure created in step 1 and calculate the maximum, minimum, and mean of the associated numerical values for each unique company.


**Reasoning**:
Import the defaultdict class and create the dictionary to store values per company.



```python
from collections import defaultdict

company_values = defaultdict(list)

for company, valor in company_data:
    company_values[company].append(valor)
```

**Reasoning**:
Calculate the max, min, and mean for each company and store them in a dictionary.



```python
import statistics

company_stats = {}

for company, values in company_values.items():
    max_value = max(values)
    min_value = min(values)
    mean_value = statistics.mean(values)
    company_stats[company] = {'max': max_value, 'min': min_value, 'mean': mean_value}

print("\nCompany Statistics (Max, Min, Mean Valor Actual):")
for company, stats in company_stats.items():
    print(f"{company}: {stats}")
```

## Display results

### Subtask:
Present the calculated maximum, minimum, and mean for each company in a clear and readable format.


**Reasoning**:
Iterate through the company_stats dictionary and print the statistics for each company in a formatted way.



```python
print("\n--- Company Performance Summary ---")
for company, stats in company_stats.items():
    print(f"Company: {company}")
    print(f"  Maximum Valor Actual: {stats['max']:.2f}")
    print(f"  Minimum Valor Actual: {stats['min']:.2f}")
    print(f"  Average Valor Actual: {stats['mean']:.2f}")
    print("-" * 20)
```
