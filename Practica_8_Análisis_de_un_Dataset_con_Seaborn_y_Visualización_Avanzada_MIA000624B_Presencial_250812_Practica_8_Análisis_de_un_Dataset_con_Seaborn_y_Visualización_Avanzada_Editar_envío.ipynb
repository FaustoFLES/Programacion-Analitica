```python
import seaborn as sns
df = sns.load_dataset('tips')
display(df.head())
display(df.info())
display(df.describe())

display(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

# Select only the numerical columns
df_numeric = df[['total_bill', 'tip', 'size']]

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='tip', data=df, palette='viridis')
plt.title('Distribucin de Propinas por Da')
plt.xlabel('Da de la Semana')
plt.ylabel('Propina')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day', palette='viridis')
plt.title('Relacin entre Total de la Cuenta y Propina por Da')
plt.xlabel('Total de la Cuenta')
plt.ylabel('Propina')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='tip', data=df, palette='viridis')
plt.title('Promedio de Propinas por Gnero')
plt.xlabel('Gnero')
plt.ylabel('Promedio de Propina')
plt.show()
```
