```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/titanic.csv')
```

```python
survival_proportion = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
print(survival_proportion)
```

```python
highest_survival = survival_proportion.idxmax()
highest_rate = survival_proportion.max()

print(f"The combination with the highest survival rate is {highest_survival} with a rate of {highest_rate:.2f}")
```

```python
lowest_survival = survival_proportion.idxmin()
lowest_rate = survival_proportion.min()

print(f"The combination with the lowest survival rate is {lowest_survival} with a rate of {lowest_rate:.2f}")
```

```python
df['FamilySize'] = df['Siblings/Spouses Aboard'] + df['Parents/Children Aboard']
display(df.head())
```

```python
df['IsLargeFamily'] = df['FamilySize'] > 3
display(df.head())
```

```python
large_family_passengers = df[df['IsLargeFamily'] == True].shape[0]
print(f"Number of passengers in large families: {large_family_passengers}")
```

```python
large_family_survival_proportion = df[df['IsLargeFamily'] == True]['Survived'].mean()
print(f"Proportion of survival for large families: {large_family_survival_proportion:.2f}")
```

```python
def categorize_age(age):
    if age < 18:
        return 'Menor de Edad'
    else:
        return 'Mayor de Edad'

df['AgeCategory'] = df['Age'].apply(categorize_age)
display(df.head())
```

```python
import numpy as np

# Calculate the average of 'Age' ignoring null values
average_age = np.nanmean(df['Age'])

# Calculate the average of 'Fare' ignoring null values
average_fare = np.nanmean(df['Fare'])

print(f"Average Age: {average_age:.2f}")
print(f"Average Fare: {average_fare:.2f}")
```

```python
# Calculate the average of 'Age' using Pandas
pandas_average_age = df['Age'].mean()

# Calculate the average of 'Fare' using Pandas
pandas_average_fare = df['Fare'].mean()

print("NumPy Averages:")
print(f"Average Age: {average_age:.2f}")
print(f"Average Fare: {average_fare:.2f}")

print("\nPandas Averages:")
print(f"Average Age: {pandas_average_age:.2f}")
print(f"Average Fare: {pandas_average_fare:.2f}")
```

```python
# Check if the average age values are close
age_consistent = np.isclose(average_age, pandas_average_age)

# Check if the average fare values are close
fare_consistent = np.isclose(average_fare, pandas_average_fare)

if age_consistent and fare_consistent:
    print("The average Age and Fare calculated by NumPy and Pandas are consistent.")
elif age_consistent:
    print("The average Age calculated by NumPy and Pandas is consistent, but the average Fare is not.")
elif fare_consistent:
    print("The average Fare calculated by NumPy and Pandas is consistent, but the average Age is not.")
else:
    print("The average Age and Fare calculated by NumPy and Pandas are not consistent.")
```

```python
# Find the minimum and maximum values of the 'Fare' column
min_fare = df['Fare'].min()
max_fare = df['Fare'].max()

# Create 5 equidistant intervals using numpy.linspace
fare_intervals = np.linspace(min_fare, max_fare, 6)

print("Fare intervals:")
print(fare_intervals)
```

```python
# Label the intervals
fare_labels = [f'Interval_{i+1}' for i in range(len(fare_intervals)-1)]

# Create a new column assigning each passenger to their fare interval
df['FareInterval'] = pd.cut(df['Fare'], bins=fare_intervals, labels=fare_labels, include_lowest=True)

display(df.head())
```

```python
# Calculate the number of passengers in each interval
passengers_per_interval = df['FareInterval'].value_counts().sort_index()
print("Number of passengers per fare interval:")
print(passengers_per_interval)

# Calculate the proportion of survivors per interval
survival_proportion_per_interval = df.groupby('FareInterval')['Survived'].mean()
print("\nProportion of survival per fare interval:")
print(survival_proportion_per_interval)
```
