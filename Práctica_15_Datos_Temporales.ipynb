Ejercicio 01:Tipo de Datos Correcto

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/Airdata.csv')
df['DateTime'] = pd.to_datetime(df['DateTime'])
display(df.head(3))
```

```python
display(df.dtypes)
```

Ejercicio 02: Extraer componentes de fecha y tiempo

```python
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Hour'] = df['DateTime'].dt.hour
display(df.head(3))
```

Ejercicio 03: Aritm√©tica - Suma de Intervalos (timedelta)

```python
df['DateTime_plus_10d_7h_15m'] = df['DateTime'] + pd.Timedelta(days=10, hours=7, minutes=15)
display(df.head(3))
```

```python
display(df[['DateTime', 'DateTime_plus_10d_7h_15m']].head())
```

Ejercicio 04: Timestamp

```python
df['DateTime_unix'] = df['DateTime'].astype('int64') // 10**9
display(df[['DateTime', 'DateTime_unix']].head(3))
```

Ejercicio 05: Filtrado de Fechas

```python
filtered_df = df[df['DateTime'] > '2020-11-01']
display(filtered_df.head())
```
