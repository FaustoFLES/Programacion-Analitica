```python
import seaborn as sns
df = sns.load_dataset('iris')
display(df.head())
df.info()
```

```python
import matplotlib.pyplot as plt

# Group by species and calculate the mean for petal_length and petal_width
df_mean = df.groupby('species')[['petal_length', 'petal_width']].mean().reset_index()

# Melt the dataframe for easier plotting with seaborn
df_mean_melted = df_mean.melt('species', var_name='petal_dimension', value_name='mean_value')

# Create the bar plot
plt.figure(figsize=(10, 7)) # Added figsize
sns.barplot(x='species', y='mean_value', hue='petal_dimension', data=df_mean_melted, palette='viridis') # Added palette

# Add labels and title
plt.xlabel('Species')
plt.ylabel('Mean Value (cm)')
plt.title('Mean Petal Length and Width by Species')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 7)) # Added figsize
sns.histplot(data=df, x='sepal_length', bins=20, kde=True, color='skyblue') # Added color
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width by Species')
plt.show()
```

```python
# Melt the dataframe for easier plotting of multiple variables
df_melted_sepal = df.melt('species', var_name='sepal_dimension', value_name='sepal_value', value_vars=['sepal_length', 'sepal_width'])

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_value', hue='sepal_dimension', data=df_melted_sepal)

# Add labels and title
plt.xlabel('Species')
plt.ylabel('Value (cm)')
plt.title('Distribution of Sepal Length and Width by Species')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, y='petal_length', x=df.index)
plt.xlabel('Observation Index')
plt.ylabel('Petal Length (cm)')
plt.title('Petal Length over Observation Index')
plt.show()
```
