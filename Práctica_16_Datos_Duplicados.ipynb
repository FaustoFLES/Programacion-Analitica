Ejercicio 01. Datos Perdidos.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/dirtydata.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
mean_calories = df['Calories'].mean()
print(f"La media de la columna 'Calories' es: {mean_calories}")
```

```python
df['Calories'] = df['Calories'].fillna(mean_calories)
print(df.isnull().sum())
```

```python
mean_calories_after_imputation = df['Calories'].mean()
print(f"La media de la columna 'Calories' después de la imputación es: {mean_calories_after_imputation}")
```

Ejercicio 02. Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration

```python
import pandas as pd
df_dirty = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/dirtydata.csv')
df_dirty.head()
```

```python
duplicate_duration_count = df_dirty.duplicated(subset=['Duration']).sum()
print(f"Cantidad de registros duplicados en la columna 'Duration': {duplicate_duration_count}")
```

```python
duration_counts = df_dirty['Duration'].value_counts()
print("Cantidad de registros por valor único en la columna 'Duration':")
print(duration_counts)
```

```python
df_unique = df_dirty.drop_duplicates()
print(f"Dimensiones del DataFrame después de eliminar duplicados: {df_unique.shape}")
```

Ejercicio 03 Duplicidad Parcial Dos columnaa: Contar y Filtrar los registros duplicados en Pulse y MaxPulse

```python
import pandas as pd
df_dirty = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/dirtydata.csv')
df_dirty.head()
```

```python
duplicate_pulse_maxpulse_count = df_dirty.duplicated(subset=['Pulse', 'Maxpulse']).sum()
print(f"Cantidad de registros duplicados en las columnas 'Pulse' y 'Maxpulse': {duplicate_pulse_maxpulse_count}")
```

```python
pulse_maxpulse_counts = df_dirty.groupby(['Pulse', 'Maxpulse']).size().reset_index(name='count')
print("Cantidad de registros por combinación única de 'Pulse' y 'Maxpulse':")
print(pulse_maxpulse_counts)
```

```python
df_unique_last = df_dirty.drop_duplicates(keep='last')
print(f"Dimensiones del DataFrame después de eliminar duplicados (manteniendo el último): {df_unique_last.shape}")
display(df_unique_last)
```

Ejercicio 05: Cálculo de la media con duplicados y sin duplicados

```python
import pandas as pd
df_dirty = pd.read_csv('/content/drive/MyDrive/Escuela/Maestria/3er Semestre/Programacion Analitica/Class data/dirtydata.csv')
df_dirty.head()
```

```python
mean_pulse = df_dirty['Pulse'].mean()
mean_maxpulse = df_dirty['Maxpulse'].mean()

print(f"La media de la columna 'Pulse' es: {mean_pulse}")
print(f"La media de la columna 'Maxpulse' es: {mean_maxpulse}")
```

```python
df_pulse_maxpulse_unique_last = df_dirty.drop_duplicates(subset=['Pulse', 'Maxpulse'], keep='last')
print(f"Dimensiones del DataFrame después de eliminar duplicados en 'Pulse' y 'Maxpulse' (manteniendo el último): {df_pulse_maxpulse_unique_last.shape}")
display(df_pulse_maxpulse_unique_last.head())
```

```python
mean_pulse_unique = df_pulse_maxpulse_unique_last['Pulse'].mean()
mean_maxpulse_unique = df_pulse_maxpulse_unique_last['Maxpulse'].mean()

print(f"La media de la columna 'Pulse' en el DataFrame sin duplicados es: {mean_pulse_unique}")
print(f"La media de la columna 'Maxpulse' en el DataFrame sin duplicados es: {mean_maxpulse_unique}")
```
