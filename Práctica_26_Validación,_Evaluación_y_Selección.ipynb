**3. GrÃ¡fica de Errores en Entrenamiento y Prueba**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np

# Define features (X) and target (y)
X = df.drop('Grade', axis=1)
y = df['Grade']

# Define the depths to evaluate
depths = [2, 5, 10, 15, 20, 25, 30, None]

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)

# Lists to store F1-macro scores
f1_macro_scores = []

# Train and evaluate Decision Tree for each depth
for depth in depths:
    dt = DecisionTreeClassifier(max_depth=depth, random_state=42)
    scores = cross_val_score(dt, X, y, cv=skf, scoring='f1_macro')
    f1_macro_scores.append(np.mean(scores))

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(range(len(depths)), f1_macro_scores, marker='o', linestyle='-')
plt.xticks(range(len(depths)), depths)
plt.xlabel('Max Depth of Decision Tree')
plt.ylabel('Mean F1-Macro Score (Cross-Validation)')
plt.title('Decision Tree Performance (F1-Macro) vs. Max Depth')
plt.grid(True)
plt.show()
```

**4. OptimizaciÃ³n de hiperparÃ¡metros con RandomizedSearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Define features (X) and target (y) - assuming X and y are already defined
# X = df.drop('Grade', axis=1)
# y = df['Grade']

# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'criterion': ['gini', 'entropy'],
    'max_depth': np.arange(5, 51, 1),
    'min_samples_split': np.arange(2, 21, 1)
}

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)

# Initialize the Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)

# Initialize RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=dt,
    param_distributions=param_dist,
    n_iter=5,  # Number of random combinations to try
    scoring='f1_macro',
    cv=skf,
    verbose=1,
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Perform the random search
random_search.fit(X, y)

# Print the best hyperparameters and best score
print("Best hyperparameters found:")
print(random_search.best_params_)
print("\nBest F1-Macro score:")
print(random_search.best_score_)
```

**5. OptimizaciÃ³n de hiperparÃ¡metros con GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Define features (X) and target (y) - assuming X and y are already defined
# X = df.drop('Grade', axis=1)
# y = df['Grade']

# Define the parameter grid for GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': np.arange(5, 51, 1),
    'min_samples_split': np.arange(2, 21, 1)
}

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)

# Initialize the Decision Tree classifier
dt = DecisionTreeClassifier(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=dt,
    param_grid=param_grid,
    scoring='f1_macro',
    cv=skf,
    verbose=1,
    n_jobs=-1  # Use all available cores
)

# Perform the grid search
grid_search.fit(X, y)

# Print the best hyperparameters and best score
print("Best hyperparameters found:")
print(grid_search.best_params_)
print("\nBest F1-Macro score:")
print(grid_search.best_score_)
```
